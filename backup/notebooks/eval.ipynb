```python
import os 
import json
import pandas as pd

OUTPUT_DIR = '../output'

results = {}
for root, dirs, files in os.walk(OUTPUT_DIR, topdown=False):
    for name in files:
        if name.endswith('dict.json'):
            path = os.path.join(root, name)
            with open(path, 'r') as f:
                results[path] = json.load(f)
```

```python
for key, value in results.items():
    new_value = [ {'hypothesis': k, 'score': v} for k,v in value.items() ]
    results[key] = new_value

for key, value in results.items():
    results[key] = pd.DataFrame(value)
```

```python
dataset_dict = {
    "test_boun": ( "hashtag_segmentation/BOUN/Test-BOUN", "BOUN"),
    "glushkova_eng": ("hashtag_segmentation/glushkova/test_eng.csv", "glushkova"),
    "glushkova_rus": ("hashtag_segmentation/glushkova/test_rus.csv", "glushkova"),
    "stanford": ("hashtag_segmentation/Stanford/stanford_dataset.txt", "stanford")
}

import importlib.util
import sys
module_name = 'reader'
file_path = '../reader.py'
spec = importlib.util.spec_from_file_location(module_name, file_path)
module = importlib.util.module_from_spec(spec)
sys.modules[module_name] = module
spec.loader.exec_module(module)
```

```python
DATASET_DIR = '../../../datasets'
for key, value in dataset_dict.items():
    dataset = module.DatasetReader(os.path.join(DATASET_DIR,value[0]), value[1])
    dataset.read()
    dataset_dict[key] = dataset.test
```
